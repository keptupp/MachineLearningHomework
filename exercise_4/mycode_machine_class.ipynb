```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
def derivation_sigmoid(z):
    return sigmoid(z)*(1-sigmoid(z))
```

```python
def spread_forward(x,theta1,theta2):
    z1=x@theta1
    hx=sigmoid(z1)
    hx=np.insert(hx,0,np.ones(hx.shape[0]),axis=1)#5000*26
    
    z2=hx@theta2
    hx2=sigmoid(z2)#5000*10
    return hx2
```
